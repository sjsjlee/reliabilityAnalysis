### 파이썬 SurPyval 라이브러리를 사용한 신뢰성 분석

```python
#### Surpyval 라이브러리 ####

import pandas as pd
import surpyval as surv

data = pd.read_csv(r"C:\sample_data\WeibullDataSet.csv")

# 데이터 변환 리스트 생성

suspension = []

for i in data['Suspension']:
    if i == 'S' :
        suspension.append(1)
    elif i == 'F' :
        suspension.append(0)

# 변환 리스트 데이터 프레임에 추가

data['suspension'] = suspension

# 패키지에 변수 넣어서 확인

my_weibull = surv.Weibull.fit(data['runDay'], data['suspension'])

# 결과 확인

my_weibull
```

```python
#### surpyval plot 그리기 -> (.plot()) 사용 ####

my_weibull.plot()
```
