### 파이썬 Weibull 라이브러리를 사용한 신뢰성 분석

### 사용한 라이브러리 소개

- __weibull__ : 일반적인 weibull 분포가 아닌 베이즈 분포를 적용한 와이블 분포를 계산하기 위한 라이브러리(weibull 분포 계산도 가능)
  - 해당 코드에서는 전체가 censored인 데이터 대상
  - beta 값 고정 가능
- Pandas : 데이터프레임 자료구조 전처리 라이브러리

```python
import pandas as pd
import weibull

data = pd.read_csv(r"C:sample_data\WeibayesDataSet.csv")

# 데이터프레임 -> 리스트 자료형 변환

data_censored = data['Suspension'] == 'S'
data_censored = data[data_censored]
list_censored = data_censored['runDay'].tolist()

# weibull 라이브러리의 weibayes 사용해 결과 확인

my_weibayes = weibull.Weibayes(data = list_censored, confidence_level= 0.95, beta = 3)

print(my_weibayes.beta)
print(my_weibayes.eta[0])

```

```python
### 플롯 출력

my_weibayes.plot()
```
